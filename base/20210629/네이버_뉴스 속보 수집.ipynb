### 네이버 뉴스 수집

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_page(url):
    html = requests.get(url, headers = headers).text
    soup = BeautifulSoup(html, 'html.parser')
    type06 = soup.find('ul',{'class':'type06_headline'})
    dl = type06.find_all('dl')
    
    return dl

def page_to_items(df, dl, page, dt):
    for item in dl:
        try:
            #이미지 링크 
            img = item.find('dt',{'class':'photo'}).find('img')
            df['img'].append(img['src'])
            #기사 링크  및 제목
            link = item.find('dt',{'class':''}).find('a')
            df['link'].append(link['href'])
            df['title'].append(link.text.replace('\t','').replace('\n','')[1:len(link.text)+1])
            #기사 내용
            content = item.find('dd')
            df['content'].append(content.text.replace('\t','').replace('\n','').split('…')[0])        
            df['date'].append(dt)
            df['page'].append(page)
        except:
            print('No Data')
        
    return df
```

```python
from datetime import date

headers = {'User-Agent':'Mozilla/5.0'}
dt = "{:%Y%m%d}".format(date.today())
df = {'img': [], 'link' : [], 'title': [], 'content': [], 'date': [], 'page' : []}

end_page = 2
end_date = 2

for i in range(0, end_date):
    temp_dt = str(int(dt) - i)    
    for j in range(0, end_page):
        page = str(j+1).zfill(3)
        url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=' + page + '&date=' + temp_dt #네이버-뉴스-속보
        dl = get_page(url)
        df = page_to_items(df, dl, page, temp_dt)
        
df = pd.DataFrame(df, columns = ['date', 'page', 'img', 'link', 'title', 'content'])
df.to_csv('df_csv.csv', index = False, encoding='utf-8')
pd.read_csv('df_csv.csv', encoding='utf-8')
```
