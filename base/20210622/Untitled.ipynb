```python
# 실습
# 1. winequality-red.csv를 읽어서 red_df 데이터 프레임 만들기
# 2. winequality-white.csv를 읽어서 white_df 데이터 프레임 만들기
```

```python
red_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
white_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
red_df = pd.read_csv(red_url, sep=";")
white_df = pd.read_csv(white_url, sep=";")
red_df
```

```python
red_df.to_numpy()
```

```python
red_df.sort_values(by="pH", ascending=False).head(5)[['density', 'pH']]
```

```python
red_df.quality.value_counts()
```

```python
new_red = red_df[['pH', 'alcohol', 'density', 'quality']]
new_red
```

```python
q1 = red_df["pH"].quantile(0.25)
q3 = red_df["pH"].quantile(0.75)
irq = q3 - q1
low =  q1 - 1.5 * irq
high = q3 + 1.5 * irq
count = 0
for i in new_red["pH"]:
    if (i < low or i > high):
        count += 1
print(count)
```

```python
new_red[(new_red["pH"] < low) | (new_red["pH"] > high)].shape[0]
```
