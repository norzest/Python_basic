pandas 집계함수 

```python
import pandas as pd
cctv_seoul = pd.read_csv("data/CCTV_in_Seoul.csv")

cctv_seoul.head(5)     #5개 행만
```

```python
print(cctv_seoul["소계"].mean() )        # "소계" 열 값의 평균
print(cctv_seoul.소계.sum() )         # "소계" 열 값의 합
print(cctv_seoul.소계.min())         # "소계" 열 값의 최소
print(cctv_seoul.소계.max())         # "소계" 열 값의 최대
print(cctv_seoul.소계.std() )         # "소계" 열 값의 표준편차
print(cctv_seoul.소계.count() )       # "소계" 열 값의 건수
```

```python
cctv_seoul.describe()
```

```python
crime_seoul = pd.read_csv ("data/crime_in_Seoul_include_gu_name.csv")


#컬럼명에 공백이 포함되면 crime_seoul.폭력 발생   <-- 이렇게 쓸수 없음. 아래처럼 사용.
print(  crime_seoul["폭력 발생"]#TODO )                   # "폭력 발생" 열 값의 평균

print( crime_seoul.#TODO ("구별")["폭력 발생"].#TODO ())     # "폭력 발생" 열 값의 평균.   단, 구별
print( crime_seoul.#TODO ("구별")["폭력 발생"].#TODO ())      # "폭력 발생" 열 값의 합.     단, 구별
print( crime_seoul.#TODO ("구별")["폭력 발생"].#TODO ())      # "폭력 발생" 열 값의 최소값.  단, 구별
print( crime_seoul.groupby("구별")["폭력 발생"].#TODO ())      # "폭력 발생" 열 값의 최대값.  단, 구별
print( crime_seoul.groupby("구별")["폭력 발생"].#TODO ())      # "폭력 발생" 열 값의 표준편차. 단, 구별
print( crime_seoul.groupby("구별")["폭력 발생"].#TODO ())    # "폭력 발생" 열 값이 존재하는 행 갯수.  단, 구별
```
