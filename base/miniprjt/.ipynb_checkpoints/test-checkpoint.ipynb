```python
night = pd.read_csv("data/nightlife_crime.csv", encoding='cp949')
night['murder'] = night['robbery'] + night['forcible_rape'] + night['thef'] + night['violence']
night['murder_tot'] = night['robbery_Tot'] + night['forcible_rape_Tot'] + night['thef_tot'] + night['violence_Tot']
night.rename(columns={'murder': 'crime'}, inplace=True)
night.rename(columns={'murder_tot': 'crime_tot'}, inplace=True)
night = night.drop('pub', axis=1)
night = night.drop('karaoke', axis=1)
night = night.drop('robbery', axis=1)
night = night.drop('forcible_rape', axis=1)
night = night.drop('thef', axis=1)
night = night.drop('violence', axis=1)
night = night.drop('robbery_Tot', axis=1)
night = night.drop('forcible_rape_Tot', axis=1)
night = night.drop('thef_tot', axis=1)
night = night.drop('violence_Tot', axis=1)
night
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

htmap = sns.heatmap(data = night.corr(), annot=True,  fmt = '.2f', linewidths=.5, cmap='Blues')
htmap
```
