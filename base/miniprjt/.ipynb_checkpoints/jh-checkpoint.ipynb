```python
x_data = df1[['violence','forcible_rape','murder','robbery']]
target = df1[['thief']]
x_data1 = sm.add_constant(x_data, has_constant='add')

multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
# R-squared 0.752로 데이터의 약 75퍼센트를 설명한다. 유의하다고 할 수 있음.
# F통계량의 P value가 0.05보다 작다. 유의하다.
# Thief 와 관련이 있는 범죄는 violence다.
```

```python
corr = df1.corr(method = 'pearson')
corr
# murder => 관련이 없다
# robbery => 관련성이 약하다
# forcilble_rape => thief, violence 와 양의 상관관계
# thief => violence와 강한 양의 상관관계
```

```python
night_df = pd.read_csv('data/nightlife_crime.csv',encoding='euc-kr')
```

```python
df2 = night_df.copy()
df2.columns = ['region','pub','karaoke','tot_nightlife','murder','robbery','forcible_rape','thief','violence','murder_tot','robbery_tot','forcible_rape_tot','thief_tot','violence_tot']
```

```python
df2_tot=df2.drop(['murder_tot','robbery_tot','forcible_rape_tot','thief_tot','violence_tot'],axis=1)
```

```python
total_crime = df2_tot.murder+df2_tot.robbery+df2_tot.forcible_rape+df2_tot.thief+df2_tot.violence
df2_total = pd.concat([df2_tot,total_crime],axis=1)
df2_total.rename(columns={0: 'total_crime'}, inplace=True)
df2_total
```

```python
x_data = df2_total[['total_crime']]
target = df2_total[['pub']]
x_data1 = sm.add_constant(x_data, has_constant='add')

multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
x_data = df2_total[['total_crime']]
target = df2_total[['karaoke']]
x_data1 = sm.add_constant(x_data, has_constant='add')

multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
X = df2_total.total_crime.values
Y = df2_total.pub.values
plt.scatter(X,Y, alpha = 0.5)
plt.xlabel('total_crime')
plt.ylabel('pub')
plt.title('total_crime~pub')
plt.show()
```

```python
X = df2_total.total_crime.values
Y = df2_total.karaoke.values
plt.scatter(X,Y, alpha = 0.5)
plt.xlabel('total_crime')
plt.ylabel('karaoke')
plt.title('total_crime~karaoke')
plt.show()
```

```python
night = pd.read_csv("data/nightlife_crime.csv", encoding='cp949')
night['murder'] = night['robbery'] + night['forcible_rape'] + night['thef'] + night['violence']
night['murder_tot'] = night['robbery_Tot'] + night['forcible_rape_Tot'] + night['thef_tot'] + night['violence_Tot']
night.rename(columns={'murder': 'crime'}, inplace=True)
night.rename(columns={'murder_tot': 'crime_tot'}, inplace=True)
night = night.drop('pub', axis=1)
night = night.drop('karaoke', axis=1)
night = night.drop('robbery', axis=1)
night = night.drop('forcible_rape', axis=1)
night = night.drop('thef', axis=1)
night = night.drop('violence', axis=1)
night = night.drop('robbery_Tot', axis=1)
night = night.drop('forcible_rape_Tot', axis=1)
night = night.drop('thef_tot', axis=1)
night = night.drop('violence_Tot', axis=1)
night
```
